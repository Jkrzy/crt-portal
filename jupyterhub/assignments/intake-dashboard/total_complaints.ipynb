```python
import datetime

start = datetime.datetime.now() - datetime.timedelta(days=90)
end = datetime.datetime.now()
```

```python
%%sql totals_results <<

SELECT assigned_section as section, COUNT(*) as total
FROM cts_forms_report
GROUP BY ROLLUP (1);
```

```python
from IPython.display import display, HTML

readable_start = start.strftime("%B %d, %Y")
readable_end = end.strftime("%B %d, %Y")
totals = totals_results.DataFrame()

total_submissions = totals[totals['section'].isnull()]['total'].values[0]
for section in totals[totals['section'].notnull()]['section'].values:
    section_submissions = totals[totals['section'] == section]['total'].values[0]
    section_percent = (section_submissions / total_submissions) * 100
    
    display(HTML(f"""
    <div class="crt-portal-card dark section-{section} filter-by-section">
    
    <h3>Total Complaints</h3>
    <hr/>
    <p style="font-size: 0.75rem">
      <em>Complaints submitted between
        <strong>{readable_start}</strong> and
        <strong>{readable_end}</strong>
      </em>
    </p>
    
    <p>
      <strong style="font-size: 2.5rem">{total_submissions:,}</strong><br/>
      Total Submissions
    </p>
    
    <p>
      <strong class="accent-blue" style="font-size: 2rem">{section_submissions:,}</strong>
      <span style="font-size: .75rem">({section_percent:.1f}% of total)</span><br/>
      Submissions to {section}
    </p>
    
    </div>
    """))
```
