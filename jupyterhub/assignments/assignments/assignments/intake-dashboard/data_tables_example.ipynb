```python
%%sql totals_results <<

SELECT
    dj_number as "DJ Number",
    viewed as "Viewed",
    primary_complaint as "Primary Complaint",
    location_name as "Organization Name",
    location_state as "State",
    DATE(create_date) as "Create Date",
    count(*) as "Number of Reports (sum)",
    count(distinct COALESCE(contact_email, '') || '|' || COALESCE(contact_first_name, '') || '|' || COALESCE(contact_last_name, '')) "Number of Unique Contacts (sum)"
FROM cts_forms_report
WHERE dj_number IS NOT NULL
GROUP BY 1,2,3,4,5,6
    
    
ORDER BY "Number of Unique Contacts (sum)" DESC
LIMIT 10000
```
